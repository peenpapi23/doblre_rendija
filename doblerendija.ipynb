##Experimento doble rendija
una simulación de un experimento cuántico de interferencia, como el experimento de las rendijas cuánticas. El experimento trata de modelar el comportamiento de partículas (como fotones o electrones) cuando pasan por dos rendijas. La característica cuántica clave es que estas partículas no solo pasan por una rendija, sino que pueden estar en una superposición de estados, es decir, pasan por ambas rendijas simultáneamente, lo que genera un patrón de interferencia.

Estado inicial: El vector X representa el estado inicial de la partícula (o fotón). Al multiplicarlo por la matriz M, que modela el sistema con las rendijas, obtenemos el siguiente estado del sistema.

Multiplicación de la matriz M y el vector X: Esta operación refleja cómo la partícula evoluciona después de interactuar con las rendijas. La matriz M puede representar diferentes caminos que la partícula puede tomar, y sus elementos complejos modelan las amplitudes de probabilidad de cada camino.

El vector resultante MMX: Cada entrada en el vector MMX es una amplitud de probabilidad compleja. Los valores reales e imaginarios que ves en el resultado final representan cómo las partículas pueden interferir entre sí. El cuadrado del módulo de cada entrada del vector indica la probabilidad de que la partícula sea detectada en un determinado estado o posición.

Interferencia cuántica: La suma de las amplitudes complejas es lo que da lugar a la interferencia cuántica, un fenómeno clave que diferencia el comportamiento cuántico del clásico. En algunos estados, las amplitudes pueden reforzarse (interferencia constructiva), mientras que en otros pueden cancelarse (interferencia destructiva).

```python
# Importar la clase Lb desde el archivo lb.py


# Creación de la instancia de la clase Lb
lb = Lb()

# Conversión de números complejos a tuplas
a = (1/(2**(1/2)), 0)
b = (-1/(6**(1/2)), (1/(6**(1/2))))
c = (-1/(6**(1/2)), (-1/(6**(1/2))))
d = (1/(6**(1/2)), (-1/(6**(1/2))))

# Multiplicaciones y suma usando las funciones de la clase
MMX = [ 0 ,
        0 ,
        0 ,
        lb.multiC(a, b),
        lb.multiC(a, c),
        lb.sumaC(lb.multiC(a, d), lb.multiC(a, b)),
        lb.multiC(a, c),
        lb.multiC(a, d)
       ]

# Impresión del resultado final
print('| ', MMX[0], ' |')
print('| ', MMX[1], ' |')
print('| ', MMX[2], ' |')
print('| ', MMX[3], ' |')
print('| ', MMX[4], ' |')
print('| ', MMX[5], ' |')
print('| ', MMX[6], ' |')
print('| ', MMX[7], ' |')
```

##Resultados
Los valores 0 significan que la partícula no estará en esos estados debido a interferencia destructiva.
Los otros valores representan la probabilidad cuántica de que la partícula sea detectada en ciertos estados tras la interacción con las rendijas.
El comportamiento de los números complejos indica el fenómeno de interferencia cuántica, donde la partícula puede estar en una superposición de caminos, causando que algunos caminos refuercen o cancelen las probabilidades de detección.

```python
import math

class Lb:
    def __init__(self):
        pass

    def sumaC(self, a, b):
        real = (a[0] + b[0])
        imag = (a[1] + b[1])
        return (real, imag)

    def multiC(self, a, b):
        real = (a[0] * b[0]) - (a[1] * b[1])
        imag = (a[0] * b[1]) + (a[1] * b[0])
        return (real, imag)

    def restaC(self, a, b):
        real = (a[0] - b[0])
        imag = (a[1] - b[1])
        return (real, imag)

    def divisionC(self, a, b):
        real = ((a[0] * b[0]) + (a[1] * b[1])) / ((b[0]**2) + (b[1]**2))
        imag = ((b[0] * a[1]) - (a[0] * b[1])) / ((b[0]**2) + (b[1]**2))
        return (real, imag)

    def moduloC(self, a):
        return ((a[0]**2) + (a[1]**2))**(1/2)

    def conjugadoC(self, a):
        imag = -1 * (a[1])
        return (a[0], imag)

    def faseC(self, a):
        fase = math.atan2(a[0], a[1])
        return fase

    def converC_P(self, a):
        modulo = self.moduloC(a)
        fase = self.faseC(a)
        return (modulo, fase)

    def converP_C(self, a):
        real = a[0] * math.cos(a[1])
        imag = a[0] * math.sin(a[1])
        return (real, imag)
```
